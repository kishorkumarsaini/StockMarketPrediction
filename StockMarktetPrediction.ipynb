```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize']=(10,8)
from matplotlib import style
style.use('ggplot')
import warnings
import seaborn as sbs
warnings.filterwarnings('ignore')

start=datetime.datetime(2010,1,1)
end=datetime.datetime(2018,12,31)

df=web.DataReader("AAPL","yahoo",start,end)
df.tail()
data=pd.DataFrame(df)
#calculate the moving averge that explain the stock downtrend and uptrend

close_px=df['Adj Close']

#calculate the rolling means.Rolling  use for visualizing the trend of stock price
#window =100 means 100 days stock price
movingmean=close_px.rolling(window=100).mean()
close_px.plot(label='AAPL')
movingmean.plot(label='MovingAVG')
plt.title("Stock Price Trend")
plt.legend()
plt.show()
data
```

## Feature Engineering

```python
import math
from sklearn import preprocessing
dfreg = data.loc[:,['Adj Close','Volume']]
dfreg['HL_PCT']=(data['High']-data['Low'])/data['Close']*100
dfreg['PCT_Change']=(data['Close']-data['Open'])/data['Open']*100
dfreg.fillna(value=-99999,inplace=True)
dfreg
'''
x=dfreg.drop('Adj Close',axis=1).values
y=dfreg['Adj Close'].values
print(x.shape)
print(y.shape)
x=preprocessing.scale(x)
'''
# Drop missing value
dfreg.fillna(value=-99999, inplace=True)
# We want to separate 1 percent of the data to forecast
forecast_out = int(math.ceil(0.01 * len(dfreg)))
# Separating the label here, we want to predict the AdjClose
forecast_col = 'Adj Close'
dfreg['label'] = dfreg[forecast_col].shift(-forecast_out)
X = np.array(dfreg.drop(['label'], 1))
# Scale the X so that everyone can have the same distribution for linear regression
X = preprocessing.scale(X)
# Finally We want to find Data Series of late X and early X (train) for model generation and evaluation
X_lately = X[-forecast_out:]
X = X[:-forecast_out]
# Separate label and identify it as y
y = np.array(dfreg['label'])
y = y[:-forecast_out]

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsRegressor

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)

#linear Regression
clflreg=LinearRegression(n_jobs=-1)
clflreg.fit(x_train,y_train)
#Quardic Regression
clfQreg=make_pipeline(PolynomialFeatures(2),Ridge())
clfQreg.fit(x_train,y_train)
#Quardic Regression with Feature 3
clfQ3reg=make_pipeline(PolynomialFeatures(3),Ridge())
clfQ3reg.fit(x_train,y_train)

#knn regression
clfknn=KNeighborsRegressor(n_neighbors=2)
clfknn.fit(x_train,y_train)


#calculate the confidence
confidLR=clflreg.score(x_test,y_test)
confidQ2=clfQreg.score(x_test,y_test)
confidQ3=clfQ3reg.score(x_test,y_test)
confidknn=clfknn.score(x_test,y_test)
print("LR Confidence",confidLR)
print("Q2 Confidence",confidQ2)
print("Q3 Confidence",confidQ3)
print("KNN Confidence",confidknn)

y_predict=clfQ3reg.predict(x_test)

stock_prediction=pd.DataFrame({'Actual':y_test,'Predicted':y_predict})
graph_plot=stock_prediction.head(25)
graph_plot.plot(kind='bar')
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
